```python
x=np.arange(0,4*np.pi,0.1)## 0.1 is a step size
y=np.sin(x)
plt.title("sin wave diagrm")

#plot the point
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,5*np.pi,0.1)
ysin=np.sin(x)
ycos=np.cos(x)

# subplot to give row and colume size
plt.subplot(2,1,1)
# make a first sin wave
plt.plot(x,ysin,"g*")
plt.title("sin diagram")

#make a second cos wave
plt.subplot(2,1,2)
plt.plot(x,ycos,"b-.")
plt.title("cos diagram")
```

## pie chart

```python
labels="python","c++","c","java"
sizes=[215,130,245,210]
colors=["gold","green","red","blue"]
explode=(0.2,0,0,0) ## explode 1st slice

#plot
plt.pie(sizes,  labels=labels, explode=explode, colors=colors, autopct='%1.1f%%', shadow=True)

plt.title("pie chart")
plt.axis("equal")
plt.show()
```
